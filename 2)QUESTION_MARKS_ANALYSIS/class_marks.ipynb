```python
def assign_grade(percentage):
    if percentage >= 35:
        return 'A'
    elif percentage >= 30:
        return 'B'
    elif percentage >= 25:
        return 'C'
    elif percentage >= 20:
        return 'D'
    else:
        return 'F'

a['Grade'] = a['Total'].apply(assign_grade)
a
```
